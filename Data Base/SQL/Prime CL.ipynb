```python
import pandas as pd

from sqlalchemy import create_engine
from config import db_password
```

```python
# Read in Prime data
prime_data = pd.read_csv('Data/PRIME_data.csv')

prime_df = pd.DataFrame(prime_data)
prime_df.head()
```

```python
# get data types
prime_df.dtypes
```

```python
# average PRME for each year
avg_prime_df = prime_df.groupby("DATE").mean().reset_index()
avg_prime_df = avg_prime_df.round({"PRIME": 2})
avg_prime_df
```
