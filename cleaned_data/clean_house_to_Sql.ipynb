# HOUSING PRICE DATASET

```python
# Read in clean housing data
# file_path=(Resources/clean_house_data)
html=("https://raw.githubusercontent.com/Diana7e/Dream_Team_Final_Project/Final_Project_Diana/team_project/Final_clean_house_data.csv")
clean_housing_df=pd.read_csv(html)
clean_housing_df.head()
```

```python
clean_housing_df.dtypes
```

```python
# Create a new column with building ID
clean_housing_df['BLD_ID']=list(range(1,55029))
```

```python
# shift column 'bld_id' to first position
first_column = clean_housing_df.pop('BLD_ID')
clean_housing_df.insert(0, 'BLD_ID', first_column)
clean_housing_df
```

```python
# Adding Price per Room to DataFrame
clean_housing_df["PRICEPRM"]=clean_housing_df["PRICE"]/clean_housing_df["ROOMS"]
clean_housing_df["PRICEPRM"]=round(clean_housing_df["PRICEPRM"], 1)
clean_housing_df
```

```python
#shift column PPRM to right position
PPRM= clean_housing_df.pop('PRICEPRM')
clean_housing_df.insert(14, 'PRICEPRM', PPRM)
clean_housing_df
```

```python
#Convert saledate to datetime
clean_housing_df['SALEDATE'] = pd.to_datetime(clean_housing_df['SALEDATE'], format='%Y')
clean_housing_df['SALEDATE'].head()  
```

```python
clean_housing_df['SALEDATE']=clean_housing_df['SALEDATE'].dt.year
```

```python
clean_housing_df['SALEDATE'] = clean_housing_df['SALEDATE'].astype('int64')
```
