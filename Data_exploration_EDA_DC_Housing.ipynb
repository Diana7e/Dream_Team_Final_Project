```python
prime_data = pd.read_csv('Data/PRIME_data.csv')
```

```python
prime_df = pd.DataFrame(prime_data)
prime_df
```

```python
prime_calc = prime_df.groupby("DATE")
prime_result = prime_calc.agg({'PRIME':'mean'})
prime_result
```

```python
prime_result.reset_index(inplace=True)
prime_result
```

```python
prime_result.dtypes
```

```python
type(prime_result)
```

```python
# Combining imp_df and prime_df
imp_df.dtypes
```

```python
imp_df['SALEDATE'] = imp_df['SALEDATE'].astype(int)
imp_df.dtypes
```

```python
# Renaming column for merge
prime_result = prime_result.rename(columns={'DATE': 'SALEDATE', 'PRIME': 'PRIME'})
prime_result.columns
```

```python
merged_df = pd.merge(imp_df, prime_result, how = 'left', on='SALEDATE')
merged_df.head(20)
```

```python
merged_df['PRIME'] = merged_df['PRIME'].replace(np.nan, 3.25)
merged_df.head(10)
```

```python
sns.pairplot(merged_df[['ROOMS','BATHRM','HF_BATHRM','BEDRM','KITCHENS', 'FIREPLACES']], kind="scatter", diag_kind = 'kde', plot_kws = {'alpha': 0.33, 's': 80, 'edgecolor': 'k'}, size = 4)
plt.show()
```
